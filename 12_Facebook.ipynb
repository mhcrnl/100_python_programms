There you have it! We created a DataFrame containing the daily ticker data for a specific stock and then plotted those data. Here is the full code:

```python
import pandas as pd
import pandas_datareader as pdr
import datetime

stocks = ['FB']
start = '1/1/2016'
end = datetime.date.today()

def get_stock(stock,start,end):
     return pdr.DataReader(stock,'yahoo',start,end)['Close']
df = pd.DataFrame({n: get_stock(n, start, end) for n in stocks})
df.plot(figsize=(16,8), grid=True)
```
