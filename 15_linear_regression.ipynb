```python
import matplotlib.pyplot as plt
%matplotlib inline

x = [5,7,3,8,4,12,13,56,]
y = [45,67,89,45,34,56,78,90]

plt.scatter(x,y)
```

```python
from scipy import stats

slope, intercept, r, p, std_err = stats.linregress(x,y)

def myfunc(x):
    return slope * x + intercept

mymodel = list(map(myfunc, x))

plt.scatter(x, y)
plt.plot(x, mymodel)
```
