```python
import pandas_datareader as pdr
import datetime
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('seaborn-whitegrid')
import numpy as np

stock = "AAL"
start = datetime.datetime(2020, 1, 1)
end = datetime.date.today()

df = pdr.get_data_yahoo(stock, start, end)

print(df.head())
print(df.tail())
print(df.describe())
```

```python
fig, ax = plt.subplots(figsize=(12,8))
ax = plt.plot(df['Close'], label='Pretul la inchidere')
plt.title("American Airline Group Inc.(AAL)")
plt.ylabel("Pretul in USD.")
```

```python
def media_mobila_simpla(cost_actiune, nr_zile):
    act_mms = cost_actiune.rolling(window=nr_zile).mean()
    return act_mms

act_mms_6 = media_mobila_simpla(df['Close'], 6)
```

```python
fig, ax = plt.subplots(figsize=(12,8))
ax = plt.plot(df['Close'], label='Pretul la inchidere')
ax = plt.plot(act_mms_6, label="Media Mobila Simpla 6zile")
plt.title("American Airline Group Inc.(AAL)")
plt.ylabel("Pretul in USD.")
```

```python
act_mms_12 = media_mobila_simpla(df['Close'], 12)
```

```python
fig, ax = plt.subplots(figsize=(12,8))
ax = plt.plot(df['Close'], label='Pretul la inchidere')
ax = plt.plot(act_mms_6, label="Media Mobila Simpla 6zile")
ax = plt.plot(act_mms_12, label="Media Mobila Simpla 12zile")
plt.title("American Airline Group Inc.(AAL)")
plt.ylabel("Pretul in USD.")
```
