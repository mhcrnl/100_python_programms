```
import newspaper

cnn_paper = newspaper.build('http://cnn.com')
lemonde_paper = newspaper.build('http://www.lemonde.fr/', language='fr')
```

```
for article in cnn_paper.articles:
  print(article.url)
```

```
print(cnn_paper.size())
```

```
for article in lemonde_paper.articles:
  print(article.url)
  print(article.title)
```

```
first_article = lemonde_paper.articles[0]
print(first_article)
```

```
first_article.download()
first_article.parse()
print(first_article.text)
```
