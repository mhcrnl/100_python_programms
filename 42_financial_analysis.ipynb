```python
stock = 'GE'
start = '2020-01-01'
end = datetime.date.today()

df = yf.download(stock, start, end)
df.head(6)
```

```python
df.tail(6)
```

```python
df.describe()
```

```python
plt.plot(df['Close'])
```

```python
fig, ax = plt.subplots(figsize=(14,8))
ax = plt.plot(df['Close'], label='General Electric pretul la inchidere.')
plt.title('General Electric')
plt.ylabel('Pretul in USD.')
```

```python
def media_mobila_simpla(cost_actiune, nr_zile):
    act_mms = cost_actiune.rolling(window=nr_zile).mean()
    return act_mms

act_mms_6 = media_mobila_simpla(df['Close'], 6)
```

```python
fig, ax = plt.subplots(figsize=(14,8))
ax = plt.plot(df['Close'], label='General Electric pretul la inchidere.')
ax = plt.plot(act_mms_6, label="Media mobila la 6 zile.")
plt.title('General Electric')
plt.ylabel('Pretul in USD.')
```

```python
act_mms_12 = media_mobila_simpla(df['Close'], 12)
```

```python
fig, ax = plt.subplots(figsize=(14,8))
ax = plt.plot(df['Close'], label='General Electric pretul la inchidere.')
ax = plt.plot(act_mms_6, label="Media mobila la 6 zile.")
ax = plt.plot(act_mms_12, label="Media mobila simpla la 12 zile.")
plt.title('General Electric')
plt.ylabel('Pretul in USD.')
```
